```python
                                # Prewitt Edge Detection with Python (OpenCV)
import cv2
import numpy as np

img = cv2.imread('./cat.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gaussian = cv2.GaussianBlur(gray,(3,3),1)

# Prewitt : 고정된 필터 사용
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])

img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gaussian2 = cv2.GaussianBlur(rgb,(3,3),1)
img_prewittx2 = cv2.filter2D(img_gaussian2, -1, kernelx)
img_prewitty2 = cv2.filter2D(img_gaussian2, -1, kernely)

cv2.imshow("Original Image", img)
cv2.imshow("Prewitt", img_prewittx + img_prewitty)
cv2.imshow("Prewitt2", img_prewittx2 + img_prewitty2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
                             # Sobel Edge Detection with Python (OpenCV)
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('./dog.jpg',1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Sobel 필터
'''
<인터넷 찾아서 메모>
ddepth: 영상의 차원
dx: x방향 미분 차수 (ex. 3차 미분)
scale: 연산 결과에 곱할 값
delta:       "    더할 값
'''
edgesx = cv2.Sobel(src=img, ddepth=-1, dx=3, dy=0, 
                   ksize=11, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)
edgesy = cv2.Sobel(img, -1, dx=0, dy=3,
                   ksize=11, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)
edges = edgesx + edgesy

output = [img, edgesx, edgesy, edges]
titles = ['Original', 'dx=1 dy=0', 'dx=0 dy=1', 'Edges']
    
for i in range(4):
        plt.subplot(2, 2, i+1)
        plt.imshow(output[i], cmap = 'gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
plt.show()
```

```python
                            #Canny Edge Detection with Python (OpenCV)
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('./vision.jpg')   #Applys the OpenCV  cv2.Canny() function
edges = cv2.Canny(img, 100, 200)   # edge 검출 함수 Canny(이미지, lower threshold, upper threshold) threshold 넘으면 무시

#Displays final results
cv2.imshow('Original Image',img)
cv2.imshow('Edge Image', edges)

cv2.imwrite('./outedges111.bmp', edges)  # Press Q on keyboard to  exit
cv2.waitKey(0)
cv2.destroyAllWindows()      #Close all the frames (images)
```

```python
                         #Comparing Edge Detection Techniques using Python (OpenCV)
import cv2
import numpy as np

img = cv2.imread('./vision.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gaussian = cv2.GaussianBlur(gray,(3,3),0)

#canny
img_canny = cv2.Canny(img,100,200)

#sobel
img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely

#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)

cv2.imshow("Original Image", img)
cv2.imshow("Canny", img_canny)
cv2.imshow("Sobel X", img_sobelx)
cv2.imshow("Sobel Y", img_sobely)
cv2.imshow("Sobel", img_sobel)
cv2.imshow("Prewitt X", img_prewittx)
cv2.imshow("Prewitt Y", img_prewitty)
cv2.imshow("Prewitt", img_prewittx + img_prewitty)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
